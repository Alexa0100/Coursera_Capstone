Balancing the data: Step One seperate the majority and minority classes

```python
df_majority = df_data[df_data.SEVERITYCODE==1]
```

```python
df_minority = df_data[df_data.SEVERITYCODE==2]
```

Balancing the data: Downsample the majority class

```python
df_majority_downsampled = resample(df_majority, 
                                 n_samples=58188)
```

```python
df_downsampled = pd.concat([df_majority_downsampled, df_minority])
```

```python
df_downsampled.head()
```

```python
df_downsampled.SEVERITYCODE.value_counts()
```

```python
df_minority_upsampled = resample(df_minority, 
                                 replace=True,    
                                 n_samples=58188) 
```

```python
df_balanced = pd.concat([df_majority_downsampled, df_minority_upsampled])
```
